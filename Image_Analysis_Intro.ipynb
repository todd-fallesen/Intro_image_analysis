## Do some filters (Eddie will clean up the logic here, cause Eddie is great)

```python
Do_Rescale = True
Do_Gauss = False
Do_Median = False

sigma = 1              # kernel size for the filters
rescale_factor = 0.5   # rescale factor
image_original = image # keep a copy of original image


if Do_Rescale:
    image = rescale(image, (1,rescale_factor, rescale_factor), anti_aliasing=False)

if Do_Gauss:
    image = filters.gaussian(image, sigma=sigma)
        
if Do_Median:
    image = filters.median(image, sigma=sigma)
        

# show all test images

plt.figure(figsize=(8,5))
plt.imshow(image[image.shape[0]//2],cmap='gray')

```

```python
#check the size.  if we resized, this should be different. Did it resize in three dimensions?
image.shape
```

```python
## Now plot original image side by side with the filtered image
# Create a figure and axes

z_slice = image.shape[0]//2

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first image
axes[0].imshow(image_original[z_slice], cmap='gray')
axes[0].axis('off')  # Turn off axes
axes[0].set_title('Original Image')

# Plot the second image
axes[1].imshow(image[z_slice], cmap='gray')
axes[1].axis('off')  # Turn off axes
axes[1].set_title('Modified Image')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
